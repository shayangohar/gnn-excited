```python
import h5py
from itertools import islice

path = '../data/A_9.hdf5'

with h5py.File(path, 'r') as f:
    # Avoid list(f.keys()); sample only a few keys lazily.
    print(f'Total molecules: {len(f)}')

    first_key = next(iter(f))
    print(f'First key: {first_key}')
    print('First 5 keys:', list(islice(f.keys(), 5)))

    mol = f[first_key]
    print('\nground_state fields:', list(mol['ground_state'].keys()))
    print('excited_state fields:', list(mol['excited_state'].keys()))
```
